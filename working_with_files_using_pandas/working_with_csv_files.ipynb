# 14. Converters

```python
pd.read_csv("matches.csv")
'''  Here we are converting long-form of team names into its short forms using converters '''
```

```python
def rename(name):
  if name == 'Royal Challengers Bangalore':
    return 'RCB'
  elif name == 'Mumbai Indians':
    return 'MI'
  elif name == 'Kolkata Knight Riders':
    return 'KKR'
  elif name == 'Kings XI Punjab':
    return 'KXIP'
  elif name == 'Chennai Super Kings':
    return 'CSK'
  elif name == 'Rajasthan Royals':
    return 'RR'
  elif name == 'Delhi Daredevils':
    return 'DD'
  elif name == 'Deccan Chargers':
    return 'DC'
  elif name == 'Sunrisers Hyderabad':
    return 'SRH'
  else:
    return name
```

```python
rename('Royal Challengers Bangalore')
```

```python
# applying the converter/function  to dataframe to change the names
pd.read_csv("matches.csv",converters={"team1":rename})
```
